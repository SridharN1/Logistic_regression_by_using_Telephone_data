```python
churn_df=pd.read_csv(r'C:\Users\SRIDHAR N\Downloads\ChurnData.csv')
churn_df.head(10)
#loading data from computer memory and reading first 10 rows of data
```

```python
churn_df=churn_df[['tenure','age', 'address','income','ed','employ','equip','callcard','wireless','churn']]
churn_df['chrun']=churn_df['churn'].astype('int')
churn_df.head()
#reading the the data from row zero
#specify the the labels of data and convert the churn data as integer
```

```python
churn_df.shape
#to knowm the size of the  data
```

```python
X=np.asarray(churn_df[['tenure','age', 'address','income','ed','employ','equip','callcard','wireless']])
X[0:5]
#read only the independent (first five only)variables from given data.
```

```python
y=np.asarray(churn_df[['churn']])
y[0:5]
#reading the dependent variable from data (churn data), first five only
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
#importing the library form python for preprocesing data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)

#splitting the data by using python library as train data, and test data with size 80:20
#Print function is used  to disply the  training sets and test datas 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR=LogisticRegression(C=0.01,solver='liblinear').fit(X_train,y_train)
LR
#IMPORTING LOGISTIC REGRESSION FROM LIBRARY AND FINDING THE BEST FIT FOR GIVEN CASE
```

```python
yhat=LR.predict(X_test)
yhat

#where yhat is regression line value and here we are predicting the the churn or  not for test case 
```

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
#checking the probabilty of final class  
```
